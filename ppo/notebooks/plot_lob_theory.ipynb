```python
import matplotlib.pyplot as plt

bid_prices = [99, 98, 97] 
bid_sizes = [-15, -23, -19] 

ask_prices = [101, 102, 103]
ask_sizes = [9, 14, 29]

mid_price = 100
bid_price = bid_prices[-1]
ask_price = ask_prices[0]

fig, ax = plt.subplots(figsize=(10, 6))

ax.bar(bid_prices, bid_sizes, color="blue", edgecolor="black", width=0.5, label="Buy Limit Order")

ax.bar(ask_prices, ask_sizes, color="red", edgecolor="black", width=0.5, label="Sell Limit Order")

ax.set_xlabel("Price")
ax.set_ylabel("Depth of Market (DOM)")
ax.set_title("Schematic of an LOB")

ax.legend(loc="upper left")

ax.spines["bottom"].set_position(("data", 0))
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

xticks = ax.get_xticks()
ax.set_xticks(xticks[xticks >= 0])
xticklabels = [str(int(abs(tick))) for tick in xticks if tick >= 0]
ax.set_xticklabels(xticklabels)

yticks = ax.get_yticks()
ax.set_yticks(yticks[yticks > 0])
yticklabels = [str(int(tick)) for tick in yticks if tick > 0]
ax.set_yticklabels(yticklabels)

plt.savefig("../thesis/misc/LOB_schematic.png", format="png")

plt.tight_layout()
plt.show()
```
